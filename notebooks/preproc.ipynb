Todos os atributos são encodados usando OneHotEncoding pois não possuem relação de hierarquia

```python
preproc_clean = preproc.replace("?", pd.NA).dropna()
preproc_clean
```

```python
preproc_clean = preproc_clean.reset_index(drop=True)
```

```python
colunas_categoricas = ['workclass', 'marital-status', 'occupation', 
                       'relationship', 'race', 'gender', 'native-country']

encoder = OneHotEncoder(sparse_output=False, drop='first')

dados_onehot = encoder.fit_transform(preproc_clean[colunas_categoricas])

nomes_colunas = encoder.get_feature_names_out(colunas_categoricas)

df_onehot = pd.DataFrame(dados_onehot, columns=nomes_colunas)


colunas_restantes = ['age', 'educational-num', 'income']
df_final_sem_interrro = pd.concat([preproc_clean[colunas_restantes], df_onehot], axis=1)
```

## Escalonamento dos Dados

Escalonar os dados é importante para algoritmos como KNN e Laplace, que são sensíveis à escala das features. Aqui usamos StandardScaler para normalizar todos os atributos.


```python
from sklearn.preprocessing import StandardScaler

# Separar features (X) e target (y)
X = df_final_sem_interrro.drop('income', axis=1)
y = df_final_sem_interrro['income']

# Criar o escalonador
scaler = StandardScaler()

# Ajustar e transformar os dados
X_scaled = scaler.fit_transform(X)

# Converter para DataFrame para manter os nomes das colunas
df_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Adicionar a coluna target de volta
df_scaled['income'] = y.values

print("Dados originais - Primeiras linhas:")
print(df_final_sem_interrro.head())
print("\nDados escalonados - Primeiras linhas:")
print(df_scaled.head())
print("\nEstatísticas dos dados originais:")
print(df_final_sem_interrro.describe())
print("\nEstatísticas dos dados escalonados:")
print(df_scaled.describe())
```

```python
# Salvar os dados escalonados
df_scaled.to_csv('../data/dados.escalonados.csv', index=False)
print("Dados escalonados salvos em: ../data/dados.escalonados.csv")
```
