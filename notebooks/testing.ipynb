```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sensitivity = 1
epsilon = 0.001

noise = []

for i in range(100000):
    ruido = np.random.laplace(0, sensitivity / epsilon)
    noise.append(ruido)
```

```python
noise_arr = np.array(noise)

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(noise_arr, bins=30, density=True, alpha=0.6, color='C0', label='Empírico')

b = sensitivity / epsilon
x = np.linspace(noise_arr.min() - 1, noise_arr.max() + 1, 500)
pdf = (1 / (2 * b)) * np.exp(-np.abs(x) / b)
ax.plot(x, pdf, 'r--', linewidth=2, label=f'Laplace PDF (b={b:.2f})')

ax.set_title('Distribuição do Ruído Laplaciano')
ax.set_xlabel('Ruído')
ax.set_ylabel('Densidade')
ax.legend()
plt.tight_layout()
plt.show()
```
